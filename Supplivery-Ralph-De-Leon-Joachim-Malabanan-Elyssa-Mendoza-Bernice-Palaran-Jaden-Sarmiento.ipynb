```python
import time

import telebot
from telebot import types

TOKEN = '947851383:AAE9TQEAMejPtlG-GJkBZSnoskRZLCcjqPQ'

knownUsers = []  
userStep = {}  
multidict = {}
cart = []

commands = { 
    'help': 'Presents all of the available commands.',
    'listofsupplements': 'Presents all the available supplements.',
    'bill': 'Presents the total bill for the transaction made.',

}

hideBoard = types.ReplyKeyboardRemove()  
```

```python
def get_user_step(uid):
    if uid in userStep:
        return userStep[uid]
    else:
        knownUsers.append(uid)
        userStep[uid] = 0
        print("New user detected, who hasn't used \"/start\" yet")
        return 0
```

```python
def listener(messages):
    for m in messages:
        if m.content_type == 'text':
            print(str(m.chat.first_name) + " [" + str(m.chat.id) + "]: " + m.text)
```

```python
bot = telebot.TeleBot(TOKEN)
bot.set_update_listener(listener)  
```

```python
@bot.message_handler(commands=['start'])
def command_start(m):
    cid = m.chat.id
    if cid not in knownUsers:  
        knownUsers.append(cid)  
        userStep[cid] = 0  
        bot.send_message(cid, "Welcome to Supplivery, new customer! Let me scan you.")
        bot.send_message(cid, "Scanning complete.")
        command_help(m) 
    else:
        bot.send_message(cid, "I already know you, no need for me to scan you again!")
```

```python
@bot.message_handler(commands=['help'])
def command_help(m):
    cid = m.chat.id
    help_text = "The following commands are available: \n"
    for key in commands:
        help_text += "/" + key + ": "
        help_text += commands[key] + "\n"
    bot.send_message(cid, help_text)  
```
